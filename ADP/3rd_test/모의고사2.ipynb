### 1-2. 예측을 위한 교차검증 기법들을 설명하고 적합한 방법을 선택하고 데이터를 분할하시오.

- Hold Out : train data가 적으면 성능의 분산이 커진다. 반대로 많아지면 신뢰도가 하락한다. 보통 8:2 혹은 9:1로 데이터를 나누어 사용한다.
- KFold : 홀드아웃에 비해 훈련 데이터 분할에 덜 민감한 성능을 얻을 수 있다. 중복을 허락하지 않고 데이터를 K개의 Fold로 랜덤하게 나눈 뒤 K-1개의 폴드로 모델을 훈련하고 나머지 하나의 폴드로 성능을 평가한다. 훈련데이터가 적을 경우 폴드 수를 늘리는 것이 좋다. 회귀에서 일반적으로 사용된다.
- Loocv : 1개의 데이터만 검증에 사용하고 n-1개의 데이터를 모두 학습에 사용한다. 모든 샘플에 대해서 한번씩 test를 하기 때문에 무작위성이 존재하지 않고 더 많은 데이터를 학습에 사용한다. 하지만 컴퓨팅 자원의 소모가 크다.
- StratifiedKFold : 분류에서 사용되며 target 속성값의 개수를 동일하게 배정하여 KFold의 target 편중 문제를 방지한다. 
- GroupKfold : 각 데이터를 종속변수가 아닌 변수(그룹형 변수)를 기준으로 분할한다. 기준 변수의 지정이 필요하다. 분류의 경우는 각 클래스의 특징을 살릴 수 있다.
- TimeSeriesSplit : 훈련 데이터를 키워나가는 방식으로 일정한 시간간격이 있는 데이터의 경우 활용된다.

#### RandomForest
- 장점 : 앙상블 알고리즘 중에서 비교적 빠른 수행 속도를 보이며 다양한 영역에서 좋은 성능을 보인다. 기반 알고리즘은 결정트리로써 쉽고 직관적이다.
- 단점 : 하이퍼 파라미터가 많고 튜닝을 위한 시간이 많이 소모된다.

#### LightGBM
- 장점 : XGBoost 보다 학습에 걸리는 시간이 훨씬 적고 메모리 사용량도 상대적으로 적다. 일반 부스팅 계열과 다르게 리프 중심 트리 분할 방식을 사용하여 트리의 깊이를 최소화 하고 과적합을 방지할 수 있다.
- 단점 : 데이터의 수가 적을 경우 과대적합이 발생하기 쉽다.

```python
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import auc, roc_curve, plot_roc_curve, accuracy_score, roc_auc_score, plot_confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rf = RandomForestClassifier(random_state = 602)
rf_pred = np.zeros((test.shape[0]))
i = 0
for tr_idx, val_idx in skf.split(X, y) :
    i += 1
    print(f'{i} Fold Modeling......')
    tr_x, tr_y = X.iloc[tr_idx], y.iloc[tr_idx]
    val_x, val_y = X.iloc[val_idx], y.iloc[val_idx]
    rf.fit(tr_x, tr_y)
    pred = rf.predict(val_x)
    print(f'{i} Fold Accuracy = {accuracy_score(val_y, pred)}\n')
    sub_pred = rf.predict_proba(test)[:, 1] / 15
    rf_pred += sub_pred
```

```python
lgbm = LGBMClassifier(random_state = 602)
lgbm_pred = np.zeros((test.shape[0]))
i = 0
for tr_idx, val_idx in skf.split(X, y) :
    i += 1
    print(f'{i} Fold Modeling......')
    tr_x, tr_y = X.iloc[tr_idx], y.iloc[tr_idx]
    val_x, val_y = X.iloc[val_idx], y.iloc[val_idx]
    lgbm.fit(tr_x, tr_y)
    pred = lgbm.predict(val_x)
    print(f'{i} Fold Accuracy = {accuracy_score(val_y, pred)}\n')
    sub_pred = lgbm.predict_proba(test)[:, 1] / 15
    lgbm_pred += sub_pred
```

```python
rf_label = (rf_pred > 0.5).astype(int)
lgbm_label = (lgbm_pred > 0.5).astype(int)
```

```python
rf_submission = pd.DataFrame(rf_label, index = test.index, columns = ['RainTomorrow'])
lgbm_submission = pd.DataFrame(lgbm_label, index = test.index, columns = ['RainTomorrow'])
```

## 2. 연관분석

### 2-1. 연관분석의 측도의 종류와 의미를 설명하고 연관분석의 장단점을 서술하시오.

연관분석은 장바구니분석 혹은 서열분석으로도 불리며 기업의 DB에서 상품의 구매, 서비스 등 일련의 거래 또는 사건들 간의 규칙을 발견하기 위해 사용된다. 보통 조건과 반응의 형태(If -> Then)로 이루어진다.

연관분석의 측도는 3가지로 지지도(Support), 신뢰도(Confidence), 향상도(Lift)가 있다.

#### 지지도(Support) 
- 전체 거래 중 항목 A와 B를 동시에 포함하는 거래의 비율

#### 신뢰도(Confidence)
- 항목 A를 포함한 거래 중에서 항목 A와 항목 B가 같이 포함될 확률로 연관성의 정도를 의미한다.

#### 향상도(Lift)
- 항목 A를 구매하지 않았을 때 품목 B의 구매확률에 비해 A가 구매됐을 때 품목 B의 구매확률의 증가 비율
- 1보다 큰 경우는 우연적 기회보다 높은 확률 / 1인 경우는 독립 / 1보다 작은 경우는 우연적 기회보다 낮은 확률

### 연관분석의 장단점
#### 장점
- 탐색적 기법으로 분석 결과를 쉽게 이해할 수 있다.
- 강력한 비목적성 분석기법으로 유용하게 활용이 가능하다.
- 분석을 위한 계산이 간단하다.

#### 단점
- 품목의 수가 증가하면 분석에 필요한 계산이 기하급수적으로 늘어난다.
- 너무 세분화된 품목이 있을 경우 규칙 발견시 의미없는 분석이 될 수 있다.
- 거래량이 적은 품목은 규칙 발견시 제외되기 쉽다.

### 2-2. 고객(cust_id)과 일자(년월일)별로 구매한 코너 리스트를 기준(중복 허용x)으로 연관분석을 실시하시오.(최소 지지도 0.01 이상)

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_table('dataTransactions.tab', encoding = 'cp949')
```

```python
data['date'] = data.datetime.str[:-6]
```

```python
df = data.groupby(['custid', 'date']).corner.agg(lambda x : list(x)).reset_index()
df.head()
```

```python
df.corner = [list(set(x)) for x in df.corner]
```

```python
df.sample(5)
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(df.corner)
df = pd.DataFrame(te_array, columns = te.columns_)
```

```python
df.sample(5)
```

```python
freq_items = apriori(df, min_support = 0.01, use_colnames = True)
freq_items.sample(5)
```

유아동복과 문화완구를 같이 구매할 확률은 0.010096이다.

문화완구를 단독으로 구매할 확률은 0.044268이다.
