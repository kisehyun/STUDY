```
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
w_t = torch.Tensor([1, 2, 3.]) # 계수
```

```
X = torch.cat([torch.ones(100, 1), torch.randn(100, 2)],1) # 절편 1, x1, x2 데이터 생성
```

```
X[:5]
```

```
y = torch.mv(X, w_t) + torch.randn(100) * .5
```

```
w = torch.randn(3, requires_grad = True)
```

```
lr = .1
```

```
losses = []

for epoc in range(100) :
  w.grad = None

  y_pred = torch.mv(X, w)

  loss = torch.mean((y - y_pred) ** 2)
  loss.backward()

  w.data = w.data - lr * w.grad.data

  losses.append(loss.item())
```

```
plt.plot(losses, color = 'red')
plt.show()
```

```
from torch import nn, optim
```

```
net = nn.Linear(in_features = 3, out_features = 1, bias = False) # input 3 / output 1 / 절편 x
```

```
opt = optim.SGD(net.parameters(), lr = .1)
```

```
loss_f = nn.MSELoss()
```

```
losses = []

for e in range(100) :
  opt.zero_grad()

  y_pred = net(X)

  loss = loss_f(y_pred.view_as(y), y)

  loss.backward()

  opt.step()

  losses.append(loss.item())
```

```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
X = iris.data[: 100]
y = iris.target[: 100]
```

```
X = torch.tensor(X, dtype = torch.float32)
y = torch.tensor(y, dtype = torch.float32)
```

```
net = nn.Linear(4, 1)

loss_f = nn.BCEWithLogitsLoss()

opt = optim.SGD(net.parameters(), lr = 0.2)
```

```
losses = []

for e in range(100) :
  opt.zero_grad()

  y_pred = net(X)

  loss = loss_f(y_pred.view_as(y), y)
  loss.backward()

  opt.step()

  losses.append(loss.item())
```

```
plt.plot(losses, color = 'green')
plt.show()
```

```
result = net(X)

prob = nn.functional.sigmoid(result)

y_pred = prob  > .5
```

```
(y.byte() == y_pred.view_as(y)).sum().item()
```

```
from sklearn.datasets import load_digits
```

```
df = load_digits()
```

```
X = df.data
y = df.target
```

```
X = torch.tensor(X, dtype = torch.float32)
y = torch.tensor(y, dtype = torch.int64)
```

```
net = nn.Linear(X.size()[1], 10)
```

```
loss_f = nn.CrossEntropyLoss()
opt = optim.SGD(net.parameters(), lr = .01)
```

```
losses = []
for e in range(100) :
  opt.zero_grad()

  y_pred = net(X)

  loss = loss_f(y_pred, y)
  loss.backward()

  opt.step()

  losses.append(loss.item())
```

```
plt.plot(losses)
```

```
_, y_pred = torch.max(net(X), 1)
```
